# Pyplot tutorial

https://matplotlib.org/users/pyplot_tutorial.html

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

* When you provide a single list of numbers, Pyplot assumes they're y-values and automatically provides x-values.
* Can pass x-values as the second argument:

```python
plt.plot([10, 20, 30, 40], [1, 2, 3, 4])
plt.show()
```

* Third argument is a MATLAB colour string:

```python
plt.plot([10, 20, 30, 40], [10, 5, 1, 5], 'r-')
plt.show()
```

* Axis command takes values ``xmin``, ``xmax``, ``ymin``, ``ymax``.

```python
import numpy as np
import matplotlib.pyplot as plt

# Evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

* Lines attributes that can be set like linewidth, dash style, antialiased

```python
plt.plot([1, 2, 3], [20, 30, 40], linewidth=5.0)
plt.show()
```

* Plt plot returns a list of Line2D objects, which can be used to set properties:

```python
line1, = plt.plot([1, 2, 3], [2, 3, 4], '-')
line1.set_antialiased(False)
```

* ``text()`` command can add text to chart.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

mu, sigma = 180, 40
x = mu + sigma * np.random.randn(10000)

# create a histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel("Height (cm)")
plt.ylabel("Probability")
plt.title("Histogram of Male Height")
plt.text(60,  .025, r'$\mu=180,\ \ sigma=15$')
plt.axis([40, 300, 0, 0.03])
plt.grid(True)
plt.show()
```

* Can annotate the text using the ``annotate`` method.

```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = plt.plot(t, s, lw=2)

plt.annotate(
    'local max',
    xy=(2, 1), xytext=(3, 1.5),
    arrowprops=dict(facecolor='black', shrink=0.05))
plt.ylim(-2, 2)
plt.show()
```

* Can change the scale using the ``xscale`` method
